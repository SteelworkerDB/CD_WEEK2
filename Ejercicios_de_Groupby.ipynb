EJERCICIOS DE GROUPBY

---



```python
import pandas as pd

#CARGA DE DATOS

filename = "https://raw.githubusercontent.com/mGalarnyk/Tutorial_Data/master/Pandas/CalIt2.data"
df=pd.read_csv(filename, header = None)
```

```python
df.head()
```

```python
df.rename(columns={0:'Flow_ID',1:'Date',2:'Time',3:'Count'},inplace=True)
df.head()
```

1. SELECCIONAR DATOS

---



```python
#SE ELABORAN LOS FILTROS
filtro_fecha=df['Date']=='07/24/05'
filtro_flow=df['Flow_ID']==7
#SE APLICA EL FILTRO A LAS FILAS, Y A TODAS LAS COLUMNAS
df_filtrado=df.loc[filtro_fecha & filtro_flow,:]
df_filtrado.head()
```

```python
df_filt_ord=df_filtrado.sort_values(by=['Count'], ascending=False) #ORDENANDO LOS VALORES FILTRADOS DE MAYOR A MENOR VALOR EN COUNT
top_10=df_filt_ord.head(10)
```

```python
top_10=top_10[['Time','Count']] #SOLO MANTENIENDO LA VARIABLE TIME QUE ES LA QUE NOS INTERESA
top_10
```

2. APLICAR FUNCION

---



```python
#UNA OPERACIÓN
mini=top_10['Count'].min()

top_10['Count']=top_10['Count'].apply(lambda x: x-mini)
```

```python
top_10
```

3. AGRUPAR

---



```python
#SE ELABORAN LOS FILTROS
filtro_fecha=(df['Date']>='08/01/05') & (df['Date']<='08/31/05')
filtro_flow=df['Flow_ID']==7
#SE APLICA EL FILTRO A LAS FILAS, Y A TODAS LAS COLUMNAS
df_filtrado_2=df.loc[filtro_fecha & filtro_flow,:]
df_filtrado_2.head()
```

```python
grupo=df_filtrado_2.groupby('Date') #AGRUPA POR FECHA
```

```python
conteo_fecha=grupo['Count'].max().reset_index()
conteo_fecha
```
